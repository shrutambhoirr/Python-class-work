```python
# 1.	Upload a image from downloads, print type format, mode, shape of the image. Show the image with colormap.
# Convert image into Negative and Gray. 
# Show all images using subplot.  import all required libraries.
import cv2
import matplotlib.pyplot as plt
import numpy as np

# read image from downloads
img = cv2.imread('G:/Download 2/1.jpg')

# print image information
print("Type: ", type(img))
print("Shape: ", img.shape)
print("Image Mode: ", img.dtype)

# show image with colormap

plt.imshow(img)
plt.show()

# convert to negative
negative_img = 255 - img

# convert to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# show all images using subplot
fig, axs = plt.subplots(1, 3, figsize=(10, 10))
axs[0].imshow(img)
axs[0].set_title('Original Image')
axs[1].imshow(negative_img)
axs[1].set_title('Negative Image')
axs[2].imshow(gray_img, cmap='gray')
axs[2].set_title('Grayscale Image')
plt.show()
```

```python
# 7.	Create a Bus child class that inherits from the Vehicle class. 
# The default fare charge of any vehicle is seating capacity * 50. If Vehicle is Bus instance,
# we need to add an extra 10% on full fare as a maintenance charge. 
# So total fare for bus instance will become the final amount = total fare + 10% of the total fare.



class Vehicle:
    def __init__(self, seating_capacity):
        self.seating_capacity = seating_capacity

    def fare(self):
        return self.seating_capacity * 50

class Bus(Vehicle):
    def __init__(self, seating_capacity):
        super().__init__(seating_capacity)
        self.maintenance_charge = 0.1

    def fare(self):
        fare = super().fare()
        return fare + fare * self.maintenance_charge
```

```python
bus = Bus(50)
print(bus.fare())

```

```python
# 7.	Create a Bus child class that inherits from the Vehicle class.
# The default fare charge of any vehicle is seating capacity * 50.
# If Vehicle is Bus instance, we need to add an extra 10% on full fare as a maintenance charge. 
# So total fare for bus instance will become the final amount = total fare + 10% of the total fare.
class Vehicle:
    def __init__(self, name, mileage, capacity):
        self.name = name
        self.mileage = mileage
        self.capacity = capacity

    def fare(self):
        return self.capacity * 100

class Bus(Vehicle):
    
    def fare(self):
        fare_car = self.capacity * 100 
        total_fare = fare_car + (0.1 *fare_car)
        return total_fare

School_bus = Bus("School Volvo", 12, 50)
print("Total Bus fare is:", School_bus.fare())
```

```python
# 10.	Create a function that takes a string as input and capitalizes a letter 
# if its ASCII code is even and returns its lower case version if its ASCII code is odd.
def ascii_capitalize(string):
    for i in string:
        if ord(i)%2 == 0:
            string = string.replace(i,i.upper())
        else:
            string = string.replace(i, i.lower())
    print(string)

ascii_capitalize('to be or not to be')
ascii_capitalize("THE LITTLE MERMAID")
ascii_capitalize("Oh what a beautiful morning.")
     
```

```python
# 11.	Write a function that creates a dictionary with each (key, value) 
# pair being the (lower case, upper case) versions of a letter, respectively.
def mapping(lst):
    dic = dict()
    for i in range(len(lst)):
        dic[lst[i]] = lst[i].upper()
    print(dic)

mapping(['p', 's'])
mapping(["a", "b", "c"])
mapping(["a", "v", "y", "z"])
     
```

```python
# 12.	Create a function that takes three integer arguments (a, b, c) and returns the amount of integers 
# which are of equal value.
def equal(num1, num2, num3):
    equal = 0
    if num1 == num2:
        if num2 == num3:
            equal = 3
        else:
            equal = 2
    elif num1 == num3:
        equal = 2
    elif num2 == num3:
        equal = 2

    print(equal)

equal(3,4,3)
equal(1,1,1)
equal(3,4,1)
```

```python
# 13.	Create a function that takes a list of non-negative integers and strings and return a new list without the strings.
def filter_list(lst):
    return [x for x in lst if type(x) == int and x >= 0]

print(filter_list([1, 2, "a", "b"])) # [1, 2]
print(filter_list([1, "a", "b", 0, 15])) # [1, 0, 15]
```

```python
# 14.	Write a function that calculates the factorial of a number recursively.
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
print(factorial(5)) # 120
print(factorial(3)) # 6

```

```python
# 15.	Given a list of numbers, create a function which returns the list but with each elements index in the list added to itself.
# This means you add 0 to the number at index 0, add to the number at index 1, etc...
def add_indexes(lst):
    return [x + i for i,x in enumerate(lst)]
print(add_indexes([0, 0, 0, 0, 0])) # [0, 1, 2, 3, 4]
print(add_indexes([1, 2, 3, 4, 5])) # [1, 3, 5, 7, 9]
print(add_indexes([5, 4, 3, 2, 1])) # [5, 5, 5, 5, 5]
```
